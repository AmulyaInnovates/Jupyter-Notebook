```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('C191_satellite_data.csv')
df
ndf=pd.DataFrame(df,columns=["Dry Mass (Kilograms)","Official Name of Satellite"])
ndf
ndf.replace(" ",float('NaN'),inplace=True)
ndf1= ndf.dropna()
ndf1
sdf=ndf1.sort_values(by=['Dry Mass (Kilograms)'])
sdf
tv= sdf.head(5)
tv

plt.xlabel('Dry Mass (Kilograms)')
plt.ylabel("Official Name of Satellite")
plt.xticks(rotation='vertical')

plt.bar(tv['Dry Mass (Kilograms)'],tv["Official Name of Satellite"],width=0.6, color="blue")
plt.show()
```
