```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

df= pd.read_csv('disney_movies.csv')
df
locate= df.loc[df['total_gross'] > 400000000]
locate
```

```python
plt.figure(figsize=(18,9))
plt.xticks(rotation='vertical')

rgb= np.random.rand(3)
plt.scatter(locate['movie_title'],locate['total_gross'], c=[rgb], label= 'Total Gross',alpha=0.5,s=locate['total_gross']/1000000)
plt.xlabel('Movie Titles')
plt.ylabel('Total Gross')
plt.legend()
plt.show()
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

ndf= pd.read_csv('movie_update.csv')
ndf
grouped=ndf.groupby('Year')['Title'].count().reset_index()
grouped
plt.figure(figsize=(18,9))
plt.xticks(rotation='vertical')

rgb= np.random.rand(3)
plt.scatter(grouped['Year'],grouped['Title'], c=[rgb], label= 'Number of Movies',alpha=0.3,s=grouped['Title']*2)
plt.xlabel('Years')
plt.ylabel('Title')
plt.legend()
plt.show()
```
