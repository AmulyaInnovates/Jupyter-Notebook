```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

df= pd.read_csv('austin_weather.csv')

df['month']=pd.DatetimeIndex(df['Date']).month
df

grouped= df.groupby(df['month'])[['TempHighF','TempLowF']].mean().reset_index()
grouped

plt.scatter(grouped['month'],grouped['TempHighF'], label='High Temperature')
plt.scatter(grouped['month'],grouped['TempLowF'], label='Low Temperature')
plt.show()
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

daf= pd.read_csv('austin_weather.csv')
daf['month']= pd.DatetimeIndex(df['Date']).month
daf

daf.replace('-', float('NAN'), inplace=True)
dafn=daf.dropna()
dafn
dafn['HumidityHighPercent']= dafn['HumidityHighPercent'].astype(int)
dafnn=dafn.groupby('month')[['TempHighF','HumidityHighPercent']].mean().reset_index()
dafnn

plt.scatter(dafnn['month'],dafnn['TempHighF'],label='Average Temperature')
plt.scatter(dafnn['month'],dafnn['HumidityHighPercent'],alpha=0.7,label='Average Humidity')
plt.legend()
plt.show()
```
