```python
from IPython.display import Image
import pandas as pd
import matplotlib.pyplot as plt
Image(filename='titanic.jpg') 
df= pd.read_csv('Titanic.csv')
df
ndf=df.dropna()
ndf
sdf=ndf.loc[ndf['Survived']== 1]
sdf
plt.figure(figsize=(19,7))
plt.xlabel('Age')
plt.ylabel('Count of the People Survived')
plt.hist(sdf['Age'], bins=10)
plt.title('The People That Survived In Titanic Tragedy')
plt.show()
```

```python
from IPython.display import Image
import pandas as pd
import matplotlib.pyplot as plt
Image(filename='titanic.jpg') 
df= pd.read_csv('Titanic.csv')
df
ndf=df.dropna()
ndf
total_class1=ndf.loc[ndf['Pclass']==1]['Pclass'].count()
survived1=ndf.loc[(ndf['Survived'==1]) & (ndf['Pclass']==1)]
survived_in_class1=survived1['Pclass'].count()
total_class2=ndf.loc[ndf['Pclass']==2]['Pclass'].count()
survived2=ndf.loc[(ndf['Survived'==1]) & (ndf['Pclass']==2)]
survived_in_class2=survived2['Pclass'].count()
total_class3=ndf.loc[ndf['Pclass']==3]['Pclass'].count()
survived3=ndf.loc[(ndf['Survived'==1]) & (ndf['Pclass']==3)]
survived_in_class3=survived3['Pclass'].count()
total_passengers=[total_class1,total_class2,total_class3]
total_survivers=[survived1,survived2,survived3]
index=['Total People in Class 1','Total People in Class 2','Total People in Class 3']
plt.bar(index,total_passengers, bottom=total_survivers,color='blue',label='total people')
plt.bar(index,total_survivers,color='red',label='total survived')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
